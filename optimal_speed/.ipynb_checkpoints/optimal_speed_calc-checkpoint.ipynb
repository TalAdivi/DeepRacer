```python
import numpy as np
from sklearn.cluster import KMeans
import math

# ---------------------- Define helper arrays ----------------------
racing_lines_coords_radius = []
racing_lines_coords_speed = []

# ---------------------- Define const variabels ----------------------
MIN_SPEED = 1
MAX_SPEED = 3

# ---------------------- Define helper methods ----------------------

# Input 3 coords [[x1,y1],[x2,y2],[x3,y3]]
# Retrun the radius
def circle_radius(coords):
    # Flatten the list and assign to variables
    x1, y1, x2, y2, x3, y3 = [i for sub in coords for i in sub]
    a = x1*(y2-y3) - y1*(x2-x3) + x2*y3 - x3*y2
    b = (x1**2+y1**2)*(y3-y2) + (x2**2+y2**2)*(y1-y3) + (x3**2+y3**2)*(y2-y1)
    c = (x1**2+y1**2)*(x2-x3) + (x2**2+y2**2)*(x3-x1) + (x3**2+y3**2)*(x1-x2)
    d = (x1**2+y1**2)*(x3*y2-x2*y3) + (x2**2+y2**2) * \
        (x1*y3-x3*y1) + (x3**2+y3**2)*(x2*y1-x1*y2)

    # In case a is zero (so radius is infinity)
    try:
        r = abs((b**2+c**2-4*a*d) / abs(4*a**2)) ** 0.5
    except:
        r = 999

    return r

# Create the radius array according set of 3 points from the racing lines
# Dont return anyting, update the racing_lines_coords_radius
def createRadiusArray():
    for index, [x,y] in enumerate(racing_lines_coords):
        if index == 0:
            racing_lines_coords_radius.append(circle_radius([[racing_lines_coords[154][0],racing_lines_coords[154][1]],[x,y],[racing_lines_coords[1][0],racing_lines_coords[1][1]]]))
        elif index == 154:
            racing_lines_coords_radius.append(circle_radius([[racing_lines_coords[153][0],racing_lines_coords[153][1]],[x,y],[racing_lines_coords[0][0],racing_lines_coords[0][1]]]))
        else:
            racing_lines_coords_radius.append(circle_radius([[racing_lines_coords[index-1][0],racing_lines_coords[index-1][1]],[x,y],[racing_lines_coords[index+1][0],racing_lines_coords[index+1][1]]]))

# Create the speed  array according racing_lines_coords_radius array and the consts of MIN_SPEED and 
def createMaxSpeedArray():
    C = MIN_SPEED / (min(racing_lines_coords_radius)**0.5)
    
    for radius in racing_lines_coords_radius:
        max_speed_result = ((C) * (radius ** 0.5))
        
        if max_speed_result > MAX_SPEED:
            max_speed_result = MAX_SPEED
        if max_speed_result < MIN_SPEED:
            max_speed_result = MIN_SPEED
            
        racing_lines_coords_speed.append(max_speed_result)


# Read all 155 track coordinates (x,y) after racing lines optimization. After 850 iterations
racing_lines_coords = [[0.58102712, 2.81034401],
       [0.58429556, 2.68967924],
       [0.5978948 , 2.56932659],
       [0.62170464, 2.45042324],
       [0.6555558 , 2.33406519],
       [0.69922991, 2.22128473],
       [0.75245791, 2.11302757],
       [0.8149174 , 2.01012967],
       [0.88622984, 1.91329448],
       [0.96595809, 1.8230715],
       [1.05360496, 1.739837],
       [1.14861321, 1.6637779 ],
       [1.25036659, 1.59487933],
       [1.35819114, 1.53291534],
       [1.47135317, 1.47743886],
       [1.58904815, 1.42776392],
       [1.71039683, 1.38296379],
       [1.83452424, 1.34199423],
       [1.96068607, 1.30390095],
       [2.08839458, 1.26804002],
       [2.2231517 , 1.23232374],
       [2.35724352, 1.19475484],
       [2.4906253 , 1.15521482],
       [2.6232595 , 1.11360486],
       [2.75510976, 1.06982968],
       [2.88611039, 1.02371369],
       [3.01626763, 0.97528105],
       [3.14561605, 0.9246366 ],
       [3.27414698, 0.87178366],
       [3.40178871, 0.81664194],
       [3.52736468, 0.75950003],
       [3.64999554, 0.70712133],
       [3.77233624, 0.65869566],
       [3.89493182, 0.61445687],
       [4.01790723, 0.57487946],
       [4.14129297, 0.5405118 ],
       [4.26508843, 0.51194662],
       [4.38927127, 0.48980971],
       [4.51379398, 0.47473785],
       [4.63857701, 0.46731862],
       [4.7635034 , 0.4680135 ],
       [4.88841883, 0.47712449],
       [5.01313584, 0.49479681],
       [5.13744033, 0.52103528],
       [5.26109931, 0.5557233 ],
       [5.38386921, 0.59864189],
       [5.50550424, 0.64948806],
       [5.62576456, 0.70789265],
       [5.74442345, 0.77343728],
       [5.86127344, 0.84567036],
       [5.97613087, 0.92412179],
       [6.08883881, 1.00831605],
       [6.1992683 , 1.09778361],
       [6.3073179 , 1.19207041],
       [6.4129119 , 1.2907453 ],
       [6.51599732, 1.39340566],
       [6.61654005, 1.49968107],
       [6.71452046, 1.6092353 ],
       [6.80992866, 1.72176685],
       [6.90275969, 1.83700823],
       [6.99300883, 1.95472419],
       [7.08066689, 2.07470923],
       [7.1657158 , 2.19678448],
       [7.24812419, 2.32079409],
       [7.32784309, 2.4466013 ],
       [7.40480152, 2.57408421],
       [7.47890216, 2.70313115],
       [7.55001675, 2.83363573],
       [7.61798171, 2.96549154],
       [7.68259374, 3.0985863 ],
       [7.74360595, 3.23279562],
       [7.80072462, 3.36797627],
       [7.85360726, 3.50395919],
       [7.90186218, 3.6405424 ],
       [7.94505035, 3.77748403],
       [7.98268969, 3.91449606],
       [8.01426252, 4.0512391 ],
       [8.03922598, 4.18731879],
       [8.05702574, 4.32228445],
       [8.06711245, 4.45563029],
       [8.06896051, 4.58679964],
       [8.06208834, 4.71519238],
       [8.0460789 , 4.84017533],
       [8.02059953, 4.96109548],
       [7.98541956, 5.07729526],
       [7.9404247 , 5.18812903],
       [7.88562714, 5.29297983],
       [7.82117065, 5.39127517],
       [7.74733037, 5.48250081],
       [7.66450734, 5.56621158],
       [7.5732183 , 5.6420385 ],
       [7.47408156, 5.70969177],
       [7.36780002, 5.76895964],
       [7.25514246, 5.81970342],
       [7.13692443, 5.86184936],
       [7.01398972, 5.89537831],
       [6.88719314, 5.92031435],
       [6.7573853 , 5.93671348],
       [6.62539944, 5.94465368],
       [6.49204007, 5.94422726],
       [6.35807332, 5.93553641],
       [6.22421807, 5.91869232],
       [6.09113752, 5.89381821],
       [5.95943059, 5.86105592],
       [5.82962272, 5.82057571],
       [5.7021562 , 5.77258823],
       [5.57738021, 5.71735772],
       [5.45554105, 5.65521517],
       [5.33677352, 5.58657034],
       [5.22109403, 5.5119216 ],
       [5.10839676, 5.43186243],
       [4.99845757, 5.34707778],
       [4.89097341, 5.25828764],
       [4.78561805, 5.16616714],
       [4.68225922, 5.07103816],
       [4.5807927 , 4.97315917],
       [4.48423127, 4.87574754],
       [4.38590263, 4.78174583],
       [4.28557394, 4.69161601],
       [4.18298847, 4.60589475],
       [4.07786706, 4.52520222],
       [3.96993653, 4.45019752],
       [3.85899528, 4.38145216],
       [3.74489481, 4.31948251],
       [3.62756967, 4.26467533],
       [3.50704818, 4.21724334],
       [3.38344682, 4.17720909],
       [3.25695994, 4.14439652],
       [3.127848  , 4.11842776],
       [2.99642498, 4.09872664],
       [2.86304564, 4.08453092],
       [2.72812386, 4.07482199],
       [2.59193728, 4.06890472],
       [2.45475383, 4.06606154],
       [2.31675601, 4.06579649],
       [2.18155269, 4.06331829],
       [2.04780686, 4.05767025],
       [1.9160569 , 4.04786584],
       [1.78687804, 4.03297057],
       [1.66081486, 4.01224159],
       [1.53852712, 3.98486532],
       [1.4206473 , 3.95023961],
       [1.30787226, 3.90781688],
       [1.20094766, 3.85714104],
       [1.10061457, 3.79792995],
       [1.0075779 , 3.73010738],
       [0.92248872, 3.65380464],
       [0.84593123, 3.56935114],
       [0.77841353, 3.47725834],
       [0.72036235, 3.37819907],
       [0.67212138, 3.27298453],
       [0.63395256, 3.16254047],
       [0.60603986, 3.04788364],
       [0.58849476, 2.93009865],
       [0.58102712, 2.81034401]]
```
